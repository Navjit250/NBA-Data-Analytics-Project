```python
data[total_cols].corr()
#including just int and floats
#not a fair representation as some players accumulate more points over time
#better representation would be something like per minute
```

```python
data_per_min = data
#lets do a groupby per total season, reduce small sample sizes, which will throw off permanenet ratios
data.groupby(['PLAYER', 'PLAYER_ID', 'Year'])[total_cols].sum()

#pull the indices into a column
data.groupby(['PLAYER', 'PLAYER_ID', 'Year'])[total_cols].sum().reset_index()


data_per_min = data.groupby(['PLAYER', 'PLAYER_ID', 'Year'])[total_cols].sum().reset_index()


#iterates through entire df to create this ratio of per minutes played
#creating ratios
for col in data_per_min.columns[4:]:
    data_per_min[col] = data_per_min[col] / data_per_min['MIN']

#shooting percentages
data_per_min['FG%'] = data_per_min['FGM'] / data_per_min['FGA']
data_per_min['3PT%'] = data_per_min['FG3M'] / data_per_min['FG3A']
data_per_min['FT%'] = data_per_min['FTM'] / data_per_min['FTA']
#percentage of attempts that are three pointers
data_per_min['FG3A%'] = data_per_min['FG3A'] / data_per_min['FGA']
#shooting efficiency
data_per_min['PTS/FGA%'] = data_per_min['PTS'] / data_per_min['FGA']
#percent of field goal that come from behind the arc
data_per_min['FG3M/FGM%'] = data_per_min['FG3M'] / data_per_min['FGM']
#free throw rate
data_per_min['FTA/FGA'] = data_per_min['FTA'] / data_per_min['FGA']
#TRUE SHOOTING PERCENTAGE, modern, more weighted to 3 pointers as indicated below
data_per_min['TRU%'] = 0.5*data_per_min['PTS'] / (data_per_min['FGA']+0.475*data_per_min['FTA'])
data_per_min['AST_TOV'] = data_per_min['AST'] / data_per_min['TOV']


```

```python
def hist_data(df=rs_df, min_MIN= 0, min_GP=0):
    return df.loc[(df['MIN']>=min_MIN) & (df['GP']>=min_GP), 'MIN']/\
    df.loc[(df['MIN']>=min_MIN) & (df['GP']>=min_GP), 'GP']
```

```python
#lets get our numbers on a per game basis
#30.25
#bin sizes change the bars
fig = go.Figure()
fig.add_trace(go.Histogram(x=hist_data(rs_df, 50, 5), histnorm='percent', name='RS', xbins={'start': 0, 'end': 46, 'size': 1}))
fig.add_trace(go.Histogram(x=hist_data(playoffs_df, 5, 1), histnorm='percent', name='Playoffs', xbins={'start': 0, 'end': 46, 'size': 1}))
fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.5)
fig.show()
```
